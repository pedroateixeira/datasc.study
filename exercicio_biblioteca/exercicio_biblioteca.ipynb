## Biblioteca

O objetivo é receber os dados de um usuário e a partir disso validar sua inscrição, gerenciar seus empréstimos de livros e calcular multas por atraso.

```python
#banco de dados simulados

usuarios = [
    {
        "nome": "João",
        "sobrenome": "Astúrias",
        "idade": 25,
        "carteira": True,
        "multa": False
    },
    {
        "nome": "Pedro",
        "sobrenome": "Teixeira",
        "idade": 26,
        "carteira": True, #testar a condição de bloqueio de carteira
        "multa": False
    },
     {
        "nome": "Tiago",
        "sobrenome": "Santos",
        "idade": 29,
        "carteira": False,
        "multa": True #testar a condição de bloqueio de multa
    },
     {
        "nome": "Pedro",
        "sobrenome": "Santos",
        "idade": 29,
        "carteira": True,
        "multa": True #testar a condição de bloqueio de multa

}]
```

```python
#Entrada dos usuários
print("-" * 50)
print("Sistema da Biblioteca".center(50))
buscar_nome = input("Digite seu primeiro nome para iniciar a consulta: ").strip().lower()
print("-" * 50)
```

Verificando se usuário está no sistema:

```python
correspondencias = [i for i in usuarios if i["nome"].lower() == buscar_nome] #filtrar a lista para achar correspondências pelo primeiro nome
```

```python
dados = None
if len(correspondencias) == 0:
  print("-" * 50)
  print("Erro: Usuário não está no sistema.")
elif len(correspondencias) == 1:
  dados = correspondencias[0] #Se tem apenas um, seleciona direto a opção no dicionário
else:
  print("-" * 50)
  print(f"Há mais de uma pessoa com o nome {buscar_nome.capitalize()}")
  buscar_sobrenome = input("Por favor, digite seu sobrenome para confirmar:").strip().lower()

#filtrar por sobrenome digitado

filtro_sobrenome = [i for i in correspondencias if i["sobrenome"].lower() == buscar_sobrenome]

if len(filtro_sobrenome) == 1:
  usuario_testado = filtro_sobrenome[0]

  print("-" * 37)
  print(f"Usuário Localizado: {usuario_testado['nome']} {usuario_testado['sobrenome']}")
  confirmacao = input("Você confirma que este é o seu perfil? (S/N): ").strip().lower()
  if confirmacao in["s", "sim"]:
    dados = usuario_testado
  else:
   print("-" * 50)
   print("Operação Cancelada. Sistema encerrado")

else:
  print("-" * 50)
  print("Sobrenome não associado ao nome")
```

Rodando as validações:

```python
if not dados["carteira"]:
        print("Acesso Negado: Você não possui uma carteira ativa da biblioteca.")
        print("Procure a administração para emitir seu documento.")

elif dados["multa"]:
        print("Empréstimo Bloqueado: Você possui multas pendentes no sistema.")
        print("Regularize seus débitos antes de solicitar um novo livro.")

else:
        print("Usuário Elegível para Empréstimos!")

        # Gerenciamento de Empréstimos (Prazo Padrão)
        prazo_dias = 14
        print(f"-> Prazo Padrão de Devolução: {prazo_dias} dias.")
        print("-" * 37)

        # Simulação de Cálculo de Multas por Atraso
        dias_atraso = 5
        taxa_diaria = 2.50
        multa_total = dias_atraso * taxa_diaria

        print(f"Simulação de Atraso ({dias_atraso} dias):")
        print(f"Valor total da multa a ser paga: R$ {multa_total:.2f}")

print("-" * 50)
```
